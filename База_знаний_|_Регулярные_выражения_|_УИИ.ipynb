# Регулярные выражения

## Модуль `re`

Функции для работы с регулярными выражениями размещены в модуле `re`.

Основные функции модуля:

`re.search(pattern, string)` – Найти в строке `string` первую строчку, подходящую под шаблон `pattern`;

`re.fullmatch(pattern, string)` – Проверить, подходит ли строка `string` под шаблон `pattern`;

`re.split(pattern, string, maxsplit=0)` – Аналог `str.split()`, только разделение происходит по подстрокам, подходящим под шаблон `pattern`;

`re.findall(pattern, string)` – Найти в строке `string` все непересекающиеся шаблоны `pattern`;

`re.finditer(pattern, string)` – Итератор всем непересекающимся шаблонам pattern в строке `string` (выдаются match-объекты);

`re.sub(pattern, repl, string, count=0)` – Заменить в строке `string` все непересекающиеся шаблоны `pattern` на `repl`;

**Регулярное выражение — это строка, задающая шаблон поиска подстрок в тексте.**

## Пример обычного поисковика по ключевому слову:

```python
import re

pattern = 'два'
string = 'раз два три четыре двадцать'

re.findall(pattern, string)
```

## Как выглядят регулярные выражения

Здесь `re.findall(pattern, string)` проверяет, подходит ли строка `string` под шаблон `pattern`;


Пример регулярного выражения:
`'\d-\d{3}-\d{3}-\d\d-\d\d'` – шаблон для номера телефона типа `'8-985-191-41-55'`

Здесь \d задаёт любую цифру, а \d{3} — задает последовательность из 3 цифр.

```python
pattern = '\d-\d{3}-\d{3}-\d\d-\d\d'
string = '8-985-191-41-55'

re.findall(pattern, string)
```

`'[-+]?\d+'` – Целое число, например (возможны ведущие нули).<br>
`[-+]?` — либо `-`, либо `+`, либо пусто.<br>
`'\d+'` — последовательность из 1 или более цифр.

```python
pattern = '[-+]?\d+'
string = '7, +17, -42, 0013 *8u'

re.findall(pattern, string)
```

`'[-+]?(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][-+]?\d+)?'` – Действительное число, возможно в экспоненциальной записи.

```python
pattern = '[-+]?(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][-+]?\d+)?'
string = '0.2, +5.45, -.4, 6e23, -3.17E-14, 7, +17, -42, 0013 *8u'

re.findall(pattern, string)
```

#### Операция «ИЛИ»

`|` – "ИЛИ". Применяется для выражений справа и слева. То есть строка подходит под шаблон, если подходит к одному из выражений `A|B`.

```python
pattern = 'морковк|св[её]кл|картошк|редиск'
string = 'морковка, свёкла, картошка, редиска'

re.findall(pattern, string)
```

```python
pattern = '(?:[0-9]|[a-f]|[A-F]){2}'
string = '01:oz:0f:b8'

re.findall(pattern, string)
```

## Пример использования всех основных функций

`re.search(pattern, string)` – Найти в строке `string` первую строчку, подходящую под шаблон `pattern`;

```python
match = re.search(r'\d\d\D\d\d', r'Телефон 123-45-67')
print(match[0] if match else 'Not found')
```

```python
match = re.search(r'\d\d\D\d\d', r'Телефон 1231212')
print(match[0] if match else 'Not found')
```

`re.fullmatch(pattern, string)` – Проверить, подходит ли строка `string` под шаблон `pattern`;

```python
match = re.fullmatch(r'\d\d\D\d\d', r'12-12')
print('YES' if match else 'NO')
```

```python
match = re.fullmatch(r'\d\d\D\d\d', r'12-12')
print('YES' if match else 'NO')
```

`re.split(pattern, string, maxsplit=0)` – Аналог `str.split()`, только разделение происходит по подстрокам, подходящим под шаблон `pattern`;

```python
print(re.split(r'\W+', 'Где, скажите мне, мои очки?!'))
```

`re.findall(pattern, string)` – Найти в строке `string` все непересекающиеся шаблоны `pattern`;

```python
print(re.findall(r'\d\d\.\d\d\.\d{4}',
                 r'Эта строка написана 16.07.2021, а могла бы и 14.07.2021'))
```

`re.finditer(pattern, string)` – Итератор всем непересекающимся шаблонам pattern в строке `string` (выдаются match-объекты);

```python
for m in re.finditer(r'\d\d\.\d\d\.\d{4}', r'Эта строка написана 19.01.2018, а могла бы и 01.09.2017'):
  print('Дата', m[0], 'начинается с позиции', m.start())
```

`re.sub(pattern, repl, string, count=0)` – Заменить в строке `string` все непересекающиеся шаблоны `pattern` на `repl`;

```python
print(re.sub(r'\d\d\.\d\d\.\d{4}',
             r'DD.MM.YYYY',
             r'Эта строка написана 19.01.2018, а могла бы и 01.09.2017'))
```

## Задача 1

Создать шаблон, под который подходят номера телефонов России.
Российский телефонный номер характеризуется следующим (упрощенно):
* Код страны: `+7` (или `8`);
* Коды мобильной связи: 800—999
* Далее 7 любых цифр
* между цифрами могут быть как пробелы, дефисы, скобки, так и их отсутствие

Примеры:
8 495 123-45-67<br>
84951234567<br>
+7(495)123-45-67<br>
+7-495-123-45-67<br>



Эта функция проверяет каждый элемент strings (все элементы должны быть типа `str`) на паттерн `pattern`.

```python
def check_strings(pattern, strings):
  max_length = max([len(s) for s in strings]) + 2
  for string in strings:
    length = max_length - len(string)
    if re.fullmatch(pattern, string):
      print('Строка {} подходит'.format(string + ' '*length))
    else:
      print('Строка {} НЕ подходит'.format(string + ' '*length))
```

```python
pattern = ''
strings = ['8 495 123-45-67',
           '84951234567',
           '+7(495)123-45-67',
           '+7-495-123-45-67',
           '+7-495-123-45-6k',
           '+7-495-123-45:67',]
check_strings(pattern, strings)
```

#### Ответ

```python
pattern = '(?:\+7|8)[ \(-]?\d{3}[ \)-]?\d{3}([ -]?\d\d){2}'
```

## Задача 2

В России применяются регистрационные знаки нескольких видов.
Общего в них то, что они состоят из цифр и букв. Причём используются только 12 букв кириллицы, имеющие графические аналоги в латинском алфавите — А, В, Е, К, М, Н, О, Р, С, Т, У и Х.

У частных легковых автомобилях номера:<br>
1 буква, 3 цифры, 2 буквы, затем 2-3 цифры с кодом региона. <br>
У такси: <br>
2 буквы, 3 цифры, затем 2-3 цифры с кодом региона. <br> Есть также и другие виды, но в этой задаче они не понадобятся.


```python
pattern = ''

strings = ['С227НА757', # легковой
           'С227НА67',  # легковой
          'КУ22777',    # такси
          'ХА226900',   # такси
          'ХА22300',    # такси
          'С227НА777',  # легковой
          'С22НА67',    # не подходит
          'Т22В7477',   # не подходит
          'М227К19У9',  # не подходит
           ]

check_strings(pattern, strings)
```

### Пример решения

```python
car = '[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}'
taxi = '[АВЕКМНОРСТУХ]{2}\d{3}\d{2,3}'
pattern = '(?:{}|{})'.format(taxi, car)

strings = ['С227НА757', # легковой
           'С227НА67',  # легковой
          'КУ22777',    # такси
          'ХА226900',   # такси
          'ХА22300',    # такси
          'С227НА777',  # легковой
          'С22НА67',    # не подходит
          'Т22В7477',   # не подходит
          'М227К19У9',  # не подходит
           ]

check_strings(pattern, strings)
```
